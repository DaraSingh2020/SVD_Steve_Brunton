```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os
import seaborn as sns
import PIL
from PIL import Image
```

```python
im = Image.open(r"C:\Users\Dara\Desktop\Dog.JPG") 
```

```python
plt.imshow(im)
plt.axis('off')
plt.show()
```

```python
X = np.mean(im, -1); # Convert RGB to grayscale

img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
print('The size of the gray scale picture is: ',X.shape)
U, S, VT = np.linalg.svd(X,full_matrices=False)
print('The shape of U, S and VT matrices are respectively: ',U.shape,S.shape,VT.shape)
S = np.diag(S)
```

```python
j = 0
for r in (5, 20, 100):
    # Construct approximate image
    Xapprox = U[:,:r] @ S[0:r,:r] @ VT[:r,:]
    plt.figure(j+1)
    j += 1
    img = plt.imshow(Xapprox)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('The number of singular values (or eigenvalues) is: ' + str(r))
    plt.show()
```

```python
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular Values')
plt.xlabel('Number of singular values')
plt.ylabel('Magnitude of the singular values')
plt.show()

plt.figure(2)
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))
plt.title('Singular Values: Cumulative Sum')
plt.xlabel('Number of singular values')
plt.ylabel('Culumative sum of singular values to the grand sum')
plt.show()
```
